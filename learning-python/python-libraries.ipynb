```python
import codecs
md5 = hashlib.md5()
md5.update('do something.'.encode('ascii'))
bytes_obj = md5.digest()
hex_str= md5.hexdigest()
```

```python
# type 'bytes' represents content with an escaped string, 
# under the hood the elements of types obj are integers range' 0-255.
print(type(bytes_obj), type(bytes_str[0]), len(bytes_str), bytes_str, bytes_str[0])

# convert a 'bytes' object to a list of integers, in which each integer represents 
# a byte with decimal integer value.
int_list = list(bytes_obj)
print(type(int_list), type(int_list[0]), len(int_list), int_list, int_list[0])

# covert a list of integers to a bytes object
repr_ints_to_bytes = bytes(int_list)
print(type(repr_ints_to_bytes),  repr_ints_to_bytes)

# convert a bytes object to a list of strings, each string represents 
# the value of a byte in hexdecimal
hex_str_repr_byte_list = [ hex(b) for b in bytes_obj ]
print(type(hex_str_repr_byte_list), type(hex_str_repr_byte_list[0]), hex_str_repr_byte_list, hex_str_repr_byte_list[0])

# convert a hex-string-represented-element list to a list with corresponding integer element.
int_from_hex_repr = [ int(h, 16) for h in hex_str_repr_list ]
print(int_from_hex_repr)

# convert a bytes object to a hex-represented-string, in which every 2 chars represents 
# the integer value of the corresponding byte of the bytes object in hexdecimal without 0x prefix.
hex_repr_string = codecs.encode(bytes_obj, 'hex').decode('ascii') 
print(type(hex_repr_string), hex_repr_string, hex_repr_string[:2])

# the same as the above one
hex_repr_string = ''.join([hs.lstrip('0x') for hs in [hex(b) for b in bytes_obj]])
print(hex_repr_string)

hex_list = [hex(i) for i in int_list]
print(type(hex_list[0]),hex_list)
```
